```python
import pandas as pd

data = pd.read_csv('/content/2019-Dec.csv')

df = pd.DataFrame(data)

df.head(10)

print(df.shape)

print(df.isnull().sum())
```

```python
print(df.isnull().sum()/df.shape[0]*100)
```

```python
print(df.isnull().sum().sum())
```

```python
print(df.isnull().sum().sum()/(df.shape[0]*df.shape[1])*100)
```

Dropping columns Because they contains more than 50% of null values

```python
df.drop(columns=["category_code"])
```

VIsualizing Data

```python
import seaborn as sb
import matplotlib.pyplot as plt
# df.fillna(0,inplace=True)
sb.heatmap(df.isnull())
plt.show()
```

```python
df.head(10)
```

Filling Null values using Backward filling.

```python
# df.drop(columns=["category_code"],inplace=True)
# df.fillna(method="bfill")
df.fillna(method="ffill",inplace=True)
df.head(20)
print(df.isnull().sum()/df.shape[0]*100)
```

Filling Mode Which Means the most Occoured Type or Value int the Column.

```python
df["event_type"].fillna(df["event_type"].mode()[0])
```

What if we want to Add mode in all Columns or object types We will use ....

```python
df.select_dtypes(include="object")
```

Filling all Columns with Object Type and

```python
for i in df.select_dtypes(include="object").columns:
  df[i].fillna(df[i].mode()[0], inplace=True)

df.head(20)
```

# Filling Data Using SKlearn library

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer(strategy="mean")

array = si.fit_transform(df[df.select_dtypes(include="float64").columns])

new_df = pd.DataFrame(array,columns = df.select_dtypes(include="float64").columns)

new_df
```

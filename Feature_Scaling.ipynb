#Feature Scaling


We use feature scaling because in a column in a dataset there are two types of Data one has max values and one has min values so the machine learning model takes the max values as it dominates the data therefore we use Feature scaling

Scaling Do not affect on outliers meaning they do not remove outliers

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data  = pd.read_csv('/content/2019-Dec.csv')
df = pd.DataFrame(data)
```

```python
df.head(10)
```

Using Standard Scaler

```python
ss = StandardScaler()

ss.fit(df[['product_id']])

ss.transform(df[['product_id']])

df['product_id']=ss.transform(df[['product_id']])
df.head()
```

normalization using minmaxscaler

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

mms.fit(df[['product_id']])

df['product_id'] = mms.transform(df[['product_id']])

df.head(10)
```

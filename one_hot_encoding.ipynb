```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

# We will be performing One Hot Encoding on event_type

```python
data = pd.read_csv('/content/2019-Dec.csv')
df = pd.DataFrame(data)
df.head(10)
```

There are no missing values

```python
df['event_type'].isnull().sum()
```

Type 1 = creating dummies

```python
en_data = df[['event_type']]
pd.get_dummies(en_data)
```

As we can see the Data is not in numerical format its still in Boolean So this Method is not much efficient.

```python
pd.get_dummies(en_data).info()
he = pd.DataFrame(pd.get_dummies(en_data))
he.head(5)
```

Here we will convert it into numerical


```python
pd.get_dummies(en_data).info()
he1 = pd.DataFrame(he).astype(int)
he1.head(10)
```

Type 2 - Encoder using Sklearn

Here We will use sklearn on hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

ohe.fit(en_data)

ar = ohe.transform(en_data).toarray()

ar = pd.DataFrame(ar, columns = ohe.get_feature_names_out(['event_type']))

ar.info()

ar.head(10)

```
